## Sobre a linguagem

### Popularidade

Python é uma linguagem de programação extremamente popular. O gráfico abaixo, obtido do [site Tiobe](https://www.tiobe.com/tiobe-index/), é um *ranking* histórico da popularidade de algumas linguagens de programação.

Python é uma linguagem relativamente simples de usar, bastante segura e com um ecossistema de código aberto muito rico. Talvez por esses motivos, a linguagem é muito usada em Ciência de Dados, Aprendizado de Máquina, Mineração de Dados e áreas afins.

### Características de Python

Python é uma linguagem:

- **Normalmente interpretada**: o código em Python é executado por uma máquina virtual, embora seja tecnicamente possível compilar o código Python para linguagem de máquina e muitas ferramentas usadas em programas Python sejam desenvolvidas em C, C++ ou outra linguagem normalmente compilada.
- **Dinâmica**: em contraste com linguagens "estáticas", tudo em Python acontece em tempo de execução.
- **Simples de usar**: a linguagem de programação Python é relativamente complexa. Ela possui uma grande quantidade de recursos. Mas os programas escritos na linguagem costumam ser mais simples.
- **Multiparadigma**: a linguagem Python suporta múltiplos paradigmas de programação. Podemos programar de maneira procedimental, orientada a objetos ou funcional.
- **Principalmente OO**: o principal paradigma de Python é a orientação a objetos. Praticamente tudo na linguagem é um objeto, até mesmo funções.

#### Impressão e leitura

### Declarações de funções

Em C, uma declaração de função segue um formato mais ou menos como o seguinte:

```
TIPO-RETORNO    NOME   (   LISTA-DE-PARAMETROS  )
{
    CORPO
}
```

O comando `return` é usado para retornar um único valor.

Para chamar uma função, a notação é mais ou menos a seguinte:

```
NOME  (   LISTA-DE-ARGUMENTOS   )
```

O vínculo entre argumento e parâmetro é sempre posicional. Isto é, o primeiro argumento passado para a função é vinculado com o primeiro parâmetro que a função recebe e assim por diante para o segundo, terceiro, quarto, quinto... etc.

Um exemplo:

```C
// Calcula o máximo divisor comum entre dois números
int mdc(int a, int b)
{
    while (a != b) {
        if (a > b) {
            a -= b;
        }
        else {
            b -= a;
        }
    }
    return a;
}

int main(void)
{
    int a, int b;
    printf("Digite dois números inteiros positivos: ");
    scanf("%d%d", &a, &b);
    printf("O mdc entre %d e %d é %d.\n", a, b, mdc(a, b));
}
```

Em Python, uma declaração de função segue mais ou menos o seguinte formato

```
def   NOME   (   LISTA-DE-PARAMETROS   ) :
    CORPO
```

O comando `return` pode retornar mais de um valor.

````python
def troca(parametro1, parametro2):
    return parametro2, parametro1
````

### Passagem de argumento por nome e por posição

Além de passar argumentos por posição, os argumentos de uma função podem ser passados por nome. Veja um exemplo:

````python
def função_complexa(nome, idade, profissão, trabalho):
    # Strings são concatenadas com o operador +
    # Uma linha muito longa pode ser quebrada em múltiplas linhas se
    # for finalizada com uma barra invertida
    return nome + " tem " + str(idade) + " anos, é " + profissão + \
        " e trabalha no " + trabalho
````

````python
função_complexa("Rafael", 40, "professor", "IComp")
````

````python
função_complexa("Codibentinho", 5, trabalho="Code Bench", profissão="programador")
````

Esse tipo de chamada é mais útil quando as funções têm valores padrões para os parâmetros. Se não houver um argumento vinculante a esse parâmetro, então o valor padrão é usado em vez disso.

Por exemplo, os argumentos podem ser vinculados por padrão com o valor `None`. Na função podemos testar se um argumento foi passado para esse parâmetro ou não.

````python
def função_complexa(nome, idade=None, profissão=None, trabalho=None):
    # Inicia uma lista que só tem o elemento nome (uma string)
    itens = [nome]
    
    if idade is not None:
        # Concatena a lista que só tem o elemento nome com o elemento abaixo
        itens += ["tem " + str(idade) + " anos"]

    # Segue concatenando...
    if profissão is None and trabalho is not None:
        itens += ["trabalha com alguma coisa no " + trabalho]
    
    if profissão is not None:
        if trabalho is None:
            itens += ["é " + profissão]
        if trabalho is not None:
            itens += ["é " + profissão + " no " + trabalho]
    
    # Monta uma frase com pontuação adequada para o número de
    # itens que não são vazios
    if len(itens) == 1:
        return itens[0] + " é um mistério"
    if len(itens) == 2:
        return " ".join(itens)
    if len(itens) == 3:
        return itens[0] + " " + itens[1] + " e " + itens[2]
    if len(itens) == 4:
        return itens[0] + " " + itens[1] + ", " + itens[2] + " e " + itens[3]
````

````python
função_complexa("Rafael", 40, trabalho="IComp")
````

````python
função_complexa("Juan", 38, "professor")
````

````python
função_complexa("Eulanda", profissão="professora")
````

````python
função_complexa("Souto")
````

````python
função_complexa("Fulano", 22, "profissional", "trabalho")
````

## Orientação a objetos

Embora Python seja uma linguagem de programação multi-paradigma, ela é fortemente orientada a objetos. Essencialmente, tudo em Python é um objeto, com algumas exceções, como o comando `if` e os laços de repetição. Até mesmo funções são objetos em Python.

### O que é orientação a objetos?

| **Paradigma procedimental** | **Paradigma OO** |
|:--|:--|
| Programas organizados em **procedimentos**, que definem ações|Os programas organizados em **objetos** que reúnem dados e ações em uma mesma entidade|
|Cada procedimento tem como meta principal exercer uma ação nos seus argumentos|Cada objeto possui um **estado interno** e um conjunto de ações para manipular esse estado|
|Programas chamam procedimentos e manipulam as variávies do programa|Objetos se comunicam trocando **mensagens**, cada qual manipulando seu estado interno|

No paradigma procedural, o programa é uma coleção de funções que chamam umas às outras, recebendo e retornando valores.

````python
print("mo", end="\n\n\n\n\n")
````

````python
5 & 2
````

````python
5 | 2
````

````python
4^7
````

Já na programação orientada a objetos, o programa é uma coleção de objetos que comunicam-se entre si trocando mensagens.

Não é apenas uma forma diferente de organizar o programa, é uma forma diferente de **pensar** ao escrever os algoritmos.

No **paradigma procedimental**, nós nos perguntamos:

- Quais são as **ações** que o programa precisa realizar?
- Quais **dados** os procedimentos manipulam para realizar essas ações?

- Na programação orientada a objetos (ou POO, para encurtar), os programas são **coleções de objetos**.
- Cada objeto é uma **entidade** que representa uma parte do programa.
  - Ao realizarmos um projeto orientado a objetos, devemos identificar quais entidades precisam ser representadas no nosso algoritmo para resolver um problema.
  - Esses objetos comunicam-se trocando mensagens, então precisamos pensar quais são as mensagens que eles precisam suportar.
  - Quais são os dados que eles precisam guardar para representar seus estados internos?
- O prorama funciona a partir da “conversa” entre os objetos.
  - Um objeto envia uma **mensagem** para um outro objeto.
  - O objeto que recebe a mensagem **ativa um método** para poder **responder** à essa mensagem (a resposta não é necessariamente um valor a ser retornado).
  - Os métodos manipulam o estado interno por meio dos **atributos** do objeto.

### Objetos e classes

A definição de um objeto é dada por um modelo chamado **classe**.

O obeto em si é uma **instância** da classe.

Algumas classes que o Python tem por padrão:

- Valores inteiros são objetos da classe `int`
- Strings são objetos da classe `str`
- Números de ponto flutuante são objetos da classe `float`
- Listas são objetos da classe `list`
- Números complexos são objetos da classe `complex`

### Atributos

Para acessar os atributos de um objeto, usa-se a seguinte notação:

```
   OBJETO   .   ATRIBUTO
```

Todas as classes numéricas possuem os atributos `real` e `imag`, que contêm a parte real e a parte imaginária do número.

### Objetos mutáveis e imutáveis

Alguns objetos em Python são imutáveis. Exemplos de objetos mutáveis são:

* `int`: números inteiros (ex.: `42` e `1_000_000_000_000_000_000_000`)
* `float`: números em ponto flutuante (ex.: `3.14`)
* `complex`: números complexos (ex.: `1j`, a unidade imaginária)
* `str`: cadeias de caracteres ou *strings* (ex.: `''`, `""`, `"\n"`, `"hello"`)
* `tuple`: duplas, triplas, quádruplas etc (ex.: `(1,0)`, `(2,2,2)`, `(3.14, "str")`, `((1,2))`)

Operações com esses objetos retornam objetos diferentes. Já objetos mutáveis podem ter seus  dados alterados. Exemplos de objetosmutáveis são:

* `list`: listas de acesso sequencial (ex.: `[]`, `[1]`, `[1,2,3]`)
* `dict`: dicionários; contêm pares chave/valor (ex.: `{"chave": "valor}"`)
* `set`: conjuntos (ex.: `{1}`, `{1,3,5}`, `{1, [2]}`)

Nunca se esqueça que *strings* são imutáveis:

### Criando uma classe

Uma classe pode ser criada com a palavra reservada **class**. No bloco que essa palavra cria, podemos definir métodos (funções).

**Todo** método precisa ter pelo menos um parâmetro. Convenciona-se que esse parâmetro seja chamado `self`. Ele é uma referência ao objeto que recebeu a mensagem que foi ativada. Em termos simples, ele tem o mesmo comportamento que o ponteiro `this` das linguagens Java e C++.

Via de regra, os métodos são definidos de acordo com as necessidades do projeto. Entretanto, existem alguns métodos próprios da linguagem que podemos usar para fazer **sobrecarga de operadores**.

Alguns exemplos:

- `__add__(self, outro)`: soma o valor do objeto que recebeu a mensagem com o valor de um outro objeto. Associado ao operador `+`.
- `__sub__(self, outro)`: soma do valor do objeto que recebeu a mensagem o valor de um outro objeto. Associado ao operador `-`.
- `__mul__(self, outro)`: multiplica o valor do objeto que recebeu a mensagem com o valor de um outro objeto. Associado ao operador `*`.
- `__floordiv__(self, outro)`: divide o valor do objeto que recebeu a mensagem com o valor de um outro objeto, retornando um valor inteiro. Associado ao operador `//`.
- `__truediv__(self, outro)`: divide o valor do objeto que recebeu a mensagem com o valor de um outro objeto, retornando um valor real. Associado ao operador `/`.

## Espaços de nomes, pacotes e módulos

Os programas em Python podem ser organizados na forma de espaços de nomes, pacotes e módulos.

Um "espaço de nomes" (namespace) é um mapa (normalmente implementado com dicionário) que associa nomes a objetos. Não existe nenhuma relação entre nomes utilizados em espaços de nomes diferentes. Por esse motivo, um mesmo nome pode existir em espaços diferentes, sem que haja conflito [[1](http://www.if.ufrgs.br/fis01069/Python.html)].

Um **módulo** é um espaço de nomes que pode conter uma biblioteca (funções, variáveis, classes etc.). Módulos permitem organizar recursos sem bagunçar o espaço de nomes atual. Por exemplo, o módulo `math` contém várias funções matemáticas para computação com números reais, ao passo que o módulo `cmath` contém funções equivalentes para números complexos.

Para utilizar um pacote, precisamos utilizar o comando `import`. Como exercício, importe o módulo `math` e use as funções `sqrt()` e `log10` para fazer algumas operações. Depois tente o mesmo com o módulo `cmath`.

Já um **pacote** é uma forma de organizar vários módulos em sub-módulos. Você irá reconhecer um pacote quando o nome do módulo parecer ter várias partes (exemplo: `sklearn.preprocessing.tests`).
